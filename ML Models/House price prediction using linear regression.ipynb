```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/896efa081ba888658eb55ae59f19ea6dd1d7ce9c/areavsprices.csv')
df
```

step 2- to perform EDA

step 3- Data Visualization  (creation of graphs )

```python
import matplotlib.pyplot as plt
#plt.plot(x-axis ,y-axis)
plt.plot(df['Area'],df['Prices'])
```

step 4-divide into input and output ,taking input as area 

input should be 2 dimentional array and output should 1 dimentional array

```python
x=df.iloc[0:6,0:1]  #input
x
```

OUTPUT is only 1 dimentional 

if u want the array to be 1D then only write the column index no

```python
y=df.iloc[:,1].values  #output
y
```

Step no 7- run regressor

PACKAGE is the collection of libraries,LinearREression -library

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Step 8 ,Fit the model

LinearRegression.fit(x,y)

```python
model.fit(x,y) 
#here we are mapping all the inputs and outputs 
#in imaginary axis we plot the i/p(x) ,o/p(y) on x-axis nd y-axis
```

```python
#step 9 predictig the model
y_pred = model.predict(x) 
#using the input values we predict the output
y_pred #predicted output
```

so now we compare the predicted output with actual output(y)


```python
y #actual output variable
```

```python
#as we can see  their is a huge differences between the predicted value nadt the actaul values
#that doesnt mean our model is inaccurate but is non linear 
#to check the linearity we use the formula 
#y = mx + C
#m is the slope ,x is the variable and C is the y intercept 
m = model.coef_#slope
m
```

```python
C = model.intercept_ # y-intercept(C)
C
```

```python
#y =mx+ C
y = m*1400 + C # here we are finding the individual prediction by the model for 1400 sqft
y
```

```python
#INDIVIDUAL PREDICTION
model.predict([[1400]])
```
